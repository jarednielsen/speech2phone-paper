```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp_results_dir = "../experiments/results/"
```

```python
rf_results = pd.read_csv(exp_results_dir + "RandomForestClassifier_2019-02-24-12-18-07/results.csv")
xgb_results = pd.read_csv(exp_results_dir + "XGBoost_2019-02-24-17-45-46/results.csv")
nn_results = pd.read_csv(exp_results_dir + "FCNN_2019-02-25-09-34-22/results.csv")
knn_results = pd.read_csv(exp_results_dir + "KNeighborsClassifier_2019-02-25-18-31-10/results.csv")
```

```python
knn_results
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
sc = ax.scatter(rf_results.n_estimators, rf_results.max_depth, c=rf_results.score, s=500)
plt.xlabel("n_estimators")
plt.ylabel("max_depth")
cb = fig.colorbar(sc, ax=ax)
cb.set_label("score")
plt.title("Accuracy vs Hyperparameters for Random Forest")
plt.show()
plt.savefig("rf.png")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
sc = ax.scatter(xgb_results.eta, xgb_results.max_depth, c=xgb_results.score, s=500)
plt.xlabel("eta")
plt.ylabel("max_depth")
cb = fig.colorbar(sc, ax=ax)
cb.set_label("score")
plt.title("Accuracy vs Hyperparameters for XGBoost")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
sc = ax.hexbin(nn_results.num_layers, nn_results.eta, C=nn_results.score, gridsize=10, yscale="log")
plt.xlabel("num_layers")
plt.ylabel("eta")
cb = fig.colorbar(sc, ax=ax)
cb.set_label("score")
plt.title("Accuracy vs Hyperparameters for Neural Network")
plt.show()
```

```python
knn_tuple = [(n, s) for n,s in zip(knn_results.n_neighbors, knn_results.score)]
knn_tuple.sort()
knn_tuple = np.stack(knn_tuple)
fig = plt.figure()
plt.plot(knn_tuple[:, 0], knn_tuple[:, 1])
plt.xlabel("n_neighbors")
plt.ylabel("score")
plt.title("Accuracy vs Hyperparameters for KNN")
plt.show()
```
